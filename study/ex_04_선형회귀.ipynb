```python
# 필요한 라이브러리 불러오기

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 데이터 생성
X = np.arange(2, 11, 2) # X값은 2부터 11까지 2씩 증가하며 들어감
# X는 공부 시간 느낌
y = np.arange(20, 101, 20) # y값은 20부터 101까지 20씩 증가하며 들어감
# y는 시험 성적 느낌
```

```python
# 위의 데이터를 이요하여 직선의 형태 확인해 보기
plt.plot(X, y, marker = 'o')

plt.grid()
plt.legend('X')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```
